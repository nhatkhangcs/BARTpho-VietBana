```python
# read excel
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('Anh_xa_phuong_ngu_v2.xlsx')

data
```

```python
# extract column
data = data[['Tiếng Việt', 'Gia Lai']]

data
```

```python
# remove NaN
data = data.dropna()
# remove duplicates
data = data.drop_duplicates()

# restore index
data = data.reset_index(drop=True)

data
```

```python
# the column 'Gia Lai' is the translation of 'Tiếng Việt', but a word in 'Tiếng Việt' can have multiple translations in 'Gia Lai'
# if so, we will split the translations into multiple rows, and make copies of the 'Tiếng Việt' column
# for example:
# Tiếng Việt | Gia Lai
#     A      |    B,C
# will be transformed into:
# Tiếng Việt | Gia Lai
#     A      |    B
#     A      |    C

dataLen = len(data['Gia Lai'])
for i in range(dataLen):
    if ',' in data['Gia Lai'].iloc[i]:
        translations = data['Gia Lai'].iloc[i].split(',')
        # remove spaces in the beginning of the translation
        # print(translations)
        for translation in translations:
            # remove spaces in the beginning of the translation
            translation = translation.strip()
            data = data._append({'Tiếng Việt': data['Tiếng Việt'].iloc[i], 'Gia Lai': translation}, ignore_index=True)
    
data
```

```python
len(data['Gia Lai'])
```

```python
dataLen = len(data['Gia Lai'])
for i in range(dataLen):
    if ',' in data['Gia Lai'].iloc[i]:
        # transform it into a NaN
        data['Gia Lai'].iloc[i] = np.nan
```

```python
# remove NaN
data = data.dropna()
data.reset_index(drop=True, inplace=True)

data
```

```python
#extract first column to a txt file
with open('dict.vi', 'w+', encoding='utf8') as f:
    for item in data['Tiếng Việt']:
        f.write("%s\n" % item)

#extract second column to a txt file
with open('dict.ba', 'w+', encoding='utf8') as f:
    for item in data['Gia Lai']:
        f.write("%s\n" % item)
```

```python
# divide these into train valid test
# train: 80%
# valid: 10%
# test: 10%

# read excel
data = pd.read_excel('GiaLaiCorupus.xlsx')

data
```

```python
# split
train, valid, test = np.split(df.sample(frac=1), [int(.8*len(data)), int(.9*len(data))])

# store train valid test into corresponding files: "train.vi" "train.ba" "valid.vi" "valid.ba" "test.vi" "test.ba"
# format is txt
# no header
# no index
# no column names
with open('GiaLai/train.vi', 'w', encoding='utf-8') as f:
    for item in train['Tiếng Việt']:
        f.write("%s" % item)
with open('GiaLai/train.ba', 'w', encoding='utf-8') as f:
    for item in train['Gia Lai']:
        f.write("%s" % item)

with open('GiaLai/valid.vi', 'w', encoding='utf-8') as f:
    for item in valid['Tiếng Việt']:
        f.write("%s" % item)
with open('GiaLai/valid.ba', 'w', encoding='utf-8') as f:
    for item in valid['Gia Lai']:
        f.write("%s" % item)

with open('GiaLai/test.vi', 'w', encoding='utf-8') as f:
    for item in test['Tiếng Việt']:
        f.write("%s" % item)
with open('GiaLai/test.ba', 'w', encoding='utf-8') as f:
    for item in test['Gia Lai']:
        f.write("%s" % item)
```
